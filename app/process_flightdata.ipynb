```python
import numpy as np
import pandas as pd
import db_connection as db
```

```python
df = db.read_db("flights")
df.drop_duplicates(subset=["filename", "datetime"], inplace=True)
df.fillna(0, inplace=True)
df["datetime"] = pd.to_datetime(df["datetime"], format="%Y-%m-%d %H:%M:%S")
df.head()
```

```python
df.describe().transpose()
```

```python
df = df[['latitude', 'longitude', 'datetime',
                'gps_altitude_m', 'distance_m', 'speed_km/s',
                'climb_m', 'climb_m(delta)', 'climb_rate_m/s',
                'glide_ratio', 'bearing', 'delta_bearing',
                'elapsed_time', 'delta_time', 'temp',
                'pressure', 'humidity', 'dew_point',
                'wind_speed', 'wind_deg']]
df.head()
```

```python
df.describe().transpose()
```

```python
def calc_zscore(df):
    mean = df.mean()
    std = df.std()
    result = (df - mean) / std
    result = np.where(abs(result) > 3, True, False)
    return result

df = df[df['gps_altitude_m'] > 0]

df_cleaned = df.copy()
zsore_columns = ['gps_altitude_m', 'distance_m', 'speed_km/s',
                'climb_m', 'climb_m(delta)', 'climb_rate_m/s',
                'glide_ratio', 'delta_time']

for col in zsore_columns:
    df_cleaned[col + "_zscore"] = calc_zscore(df_cleaned[col].copy())
    df_cleaned = df_cleaned[df_cleaned[col + "_zscore"] == False]
    df_cleaned.drop(columns=[col + "_zscore"], inplace=True)


df_cleaned = df_cleaned[df_cleaned['delta_time'] == 1]
df_cleaned.describe().transpose()
```

```python
pearson_corr = df_cleaned['climb_rate_m/s'].corr(df['climb_m'])
print(f'Pearson: {pearson_corr:.4f}')
```

```python
db.write_db(df_cleaned, "flights_cleaned")
df_cleaned[:100000].to_csv("flights_cleaned.csv", index=False)
```
